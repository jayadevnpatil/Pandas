```python
import pandas as pd

# Define a dictionary containing lists of data
people = {
    'first':['Corey','Jane','John'],
    'last': ['Schafer','Doe','Doe'],
    'email':['CoreySchafer@gmail.com','JaneDoe@gmail.com','JohnDoe@gmail.com']
}

# Create a DataFrame using the dictionary
df = pd.DataFrame(people)
df
```

```python
# Create a new column 'full name' in the DataFrame 'df' by concatenating 'first' and 'last' name columns
df['full name'] = df['first']+' '+df['last']
df
```

```python
# Drop the columns 'first' and 'last' from the DataFrame 'df' in place
df.drop(columns=['first','last'], inplace=True)
df
```

```python
# Split the 'full name' column into 'first' and 'last' names using space (' ') as the delimiter
# Assign the split results back to new 'first' and 'last' columns in the DataFrame 'df'

df[['first','last']] = df['full name'].str.split(' ', expand=True)
df
```

```python
# Append a new row to the DataFrame 'df' with the first name 'Tony'
df._append({'first':'Tony'}, ignore_index=True)
```

```python
# Define a dictionary containing lists of data
people = {
    "first":["Tony","Steve"],
    "last": ["Stark","Rogers"],
    "email":["TonyStark@avenge.com","Cap@avenge.com"]
}

# Create a new DataFrame 'df2' using the dictionary 'people'
df2 = pd.DataFrame(people)
df2
```

```python
# Append the DataFrame 'df2' to 'df' and reset the index
df = df._append(df2, ignore_index=True)
df
```

```python
# Drop the row with index 4 from the DataFrame 'df'
df.drop(index=4)
```

```python
# Drop rows from the DataFrame 'df' where the 'last' name is 'Doe'
filt = df['last'] == 'Doe'
df.drop(index=df[filt].index)
```
